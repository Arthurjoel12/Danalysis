```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set a random seed for reproducibility
np.random.seed(42)

# Sample data generation function
def generate_sample_data(num_rows=100):
    years = np.random.randint(2020, 2023, size=num_rows)
    regions = np.random.choice(["Region A", "Region B", "Region C"], size=num_rows)
    genders = np.random.choice(["Male", "Female"], size=num_rows)
    age_groups = np.random.choice(["0-14", "15-24", "25-49", "50+"], size=num_rows)
    hiv_prevalence = np.random.uniform(0.1, 10.0, size=num_rows)
    hiv_incidence = np.random.uniform(0.01, 2.0, size=num_rows)
    hiv_deaths = np.random.randint(0, 500, size=num_rows)
    antiretroviral_treatment = np.random.randint(0, 10000, size=num_rows)
    condom_use_rate = np.random.uniform(10.0, 90.0, size=num_rows)
    education_level = np.random.choice(["Low", "Medium", "High"], size=num_rows)
    income_level = np.random.choice(["Low", "Medium", "High"], size=num_rows)
    access_to_healthcare = np.random.choice(["Limited", "Moderate", "High"], size=num_rows)
    injection_drug_use = np.random.choice(["Low", "Moderate", "High"], size=num_rows)
    sexual_behavior = np.random.choice(["Low Risk", "Moderate Risk", "High Risk"], size=num_rows)
    hiv_awareness = np.random.choice(["Low", "Moderate", "High"], size=num_rows)

    data = np.column_stack((years, regions, genders, age_groups, hiv_prevalence, hiv_incidence, hiv_deaths,
                            antiretroviral_treatment, condom_use_rate, education_level, income_level,
                            access_to_healthcare, injection_drug_use, sexual_behavior, hiv_awareness))

    columns = ["Year", "Region", "Gender", "Age_Group", "HIV_Prevalence", "HIV_Incidence", "HIV_Deaths",
               "Antiretroviral_Treatment", "Condom_Use_Rate", "Education_Level", "Income_Level",
               "Access_to_Healthcare", "Injection_Drug_Use", "Sexual_Behavior", "HIV_Awareness"]

    df = pd.DataFrame(data, columns=columns)
    return df

# Create the sample dataset
sample_df = generate_sample_data()

# Save the dataset to a CSV file
sample_df.to_csv("hiv_infection_dataset.csv", index=False)

# Data Cleansing
# Convert numerical columns to appropriate data types
numerical_columns = ["Year", "HIV_Prevalence", "HIV_Incidence", "HIV_Deaths", "Antiretroviral_Treatment", "Condom_Use_Rate"]
sample_df[numerical_columns] = sample_df[numerical_columns].apply(pd.to_numeric)

# Data Analysis
# Summary statistics
summary_stats = sample_df.describe()

# Data Visualization
# Pairplot for numerical columns
sns.pairplot(sample_df[numerical_columns])
plt.show()

# Box plot for HIV_Prevalence by Gender
plt.figure(figsize=(8, 6))
sns.boxplot(x='Gender', y='HIV_Prevalence', data=sample_df)
plt.title('HIV Prevalence by Gender')
plt.xlabel('Gender')
plt.ylabel('HIV Prevalence')
plt.show()

# Histogram for HIV_Incidence
plt.figure(figsize=(8, 6))
sns.histplot(sample_df['HIV_Incidence'], bins=20, kde=True)
plt.title('Distribution of HIV Incidence')
plt.xlabel('HIV Incidence')
plt.ylabel('Frequency')
plt.show()

print("Sample dataset created and saved as 'hiv_infection_dataset.csv'.")
print("\nSummary Statistics:\n", summary_stats)


from google.colab import files
files.download("hiv_infection_dataset.csv")

```
